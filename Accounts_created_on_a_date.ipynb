```python
#IMPORTING
import pandas as pd
import csv
import os
import matplotlib.pyplot as plt
```

```python
#FILE PATH
path_parent = os.path.dirname(os.getcwd())
os.chdir(path_parent)
print(os.getcwd())
```

```python
#READING FILE
file = "/Users/USERS.csv"
followers_df = pd.read_csv(file)
followers_df.shape
```

```python
#checking the most popular join_dates
join_date = followers_df['join_date'].value_counts(ascending=False)
```

```python
#creating join_date data frame
join_date = join_date.to_frame()
```

```python
#setting a limit: how many of the top dates we want. Put the number in the ()
top_dates = join_date.head(25)
```

```python
#sorting the data
top_dates = top_dates.sort_values(by= "join_date", ascending=True)
```

```python
#plotting top join dates

top_dates.plot(kind = "barh", alpha=0.5,figsize=(8, 15))
plt.title("Most_Popular_Join_Dates")
plt.xlabel("Number of accounts created on a date")
plt.ylabel("Dates")
plt.savefig("Top_25_dates.png")
```
